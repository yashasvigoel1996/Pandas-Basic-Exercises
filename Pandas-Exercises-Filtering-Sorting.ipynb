### Step 4. How many products cost more than $10.00? 

Use `str` attribute to remove the $ sign and convert the column to proper numeric type data before filtering.


```python
chipo.head()
```

```python
temp=chipo.copy()
temp["item_price"] = temp.item_price.str.replace("$","")
temp["item_price"] = pd.to_numeric(temp.item_price)
temp = temp.loc[(temp.item_price>10),]
productCount = temp.shape
productCount[0]
```

### Step 5. What is the price of each item? 

print a data frame with only two columns `item_name` and `item_price`

```python
price=chipo.copy()
price=price.loc[:,["item_name","item_price"]]
price.drop_duplicates().reset_index(drop=True)
```

### Step 6. Sort by the name of the item

```python
sortItems = chipo.copy()
sortItems=sortItems.drop_duplicates(subset=["item_name"])
sortItems = pd.DataFrame(sortItems.item_name)
sortItems.sort_values(by=["item_name"],ascending=True).reset_index(drop=True)
```
