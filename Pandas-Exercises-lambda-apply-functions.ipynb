### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(ser):
    '''
    This method take age column of dataframe as input and return series of booleans on the basis that 
    if age is greater or equal to 18 then person is legal drinker and it's corresponding value of legal_drinker column is 
    1(True) and if person is less than 18 then person is not legal drinker and it's corresponding value of legal_drinker column
    is 0(False). 
    '''
    if(isinstance(ser, pd.Series)):
        legal_drinker=[]
        for i in ser:
            if(i>=18):
                legal_drinker.append(1)
            else:
                legal_drinker.append(0)
        return pd.Series(legal_drinker)
    else:
        print("Please enter series or dataframe column.")
```

```python
df["legal_drinker"]=majority(df.age)
df
```
