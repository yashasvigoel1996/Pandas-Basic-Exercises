### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
online_rt_groupby_sortby = online_rt.groupby(by=["Country"])[["Quantity"]].sum().add_prefix("SumOf").sort_values(by=["SumOfQuantity"],ascending=False)
online_rt_groupby_sortby
```

```python
online_rt_top_10_country = online_rt_groupby_sortby.iloc[1:11]
online_rt_top_10_country
```

```python
plt.hist(online_rt_top_10_country.SumOfQuantity,bins=10)
plt.show()
```

### Step 6. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries

```python
online_rt.head()
```

```python
top_three_countries=online_rt.copy()
top_three_countries=top_three_countries.groupby(by=["Country"])[["Quantity"]].sum().add_prefix("Total_")
top_three_countries=top_three_countries.sort_values(by=["Total_Quantity"],ascending=False)
top_three_countries
#From above dataframe we got our top 3 country(United Kingdom, Netherlands & EIRE) on the basis of Quantity.
```

```python
united_kingdom_online_rt_data=online_rt.copy()
united_kingdom_online_rt_data=united_kingdom_online_rt_data.loc[(united_kingdom_online_rt_data["Country"]=="United Kingdom"),]
united_kingdom_online_rt_data.head()
```

```python
plt.scatter(united_kingdom_online_rt_data.Quantity,united_kingdom_online_rt_data.UnitPrice)
plt.xlabel("Quantity")
plt.ylabel("UnitPrice")
plt.title("Quantity per UnitPrice by CustomerID for United Kingdom")
plt.show()
```
